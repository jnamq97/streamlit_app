```python
import os
import json
import cv2

json_path="C:/Users/User/Desktop/Bbox_1901/label.json"
with open(json_path,"r")as f:
    coco_data=json.load(f)
print(coco_data)
```

```python
# 
for image_info in coco_data["images"]:
    image_id = image_info["id"]
    image_name = image_info["file_name"]
    image_path = f"C:/Users/User/Desktop/Bbox_1901/{image_name}"
    image = cv2.imread(image_path)
    print(image)
    for annotation in coco_data["annotations"]:
        if annotation["image_id"] == image_id:
            bbox = annotation["bbox"]
            x, y, w, h = bbox
            x1, y1, x2, y2 = int(x), int(y), int(x + w), int(y + h)
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    #  
    cv2.imshow(image_name, image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import albumentations as A
transform = A.Compose([
    A.Resize(width=960, height=540),  #   
], bbox_params=A.BboxParams(format='coco', label_fields=['category_id']))
```

```python
import numpy as np
images = coco_data['images']
annotations = coco_data['annotations']
print(images)
print(annotations)
```

```python
test_path=""
image_folder="C:/Users/User/Desktop/Bbox_1901/"
```

```python
#    
for img_info in images:
    img_id = img_info['id']
    img_path = os.path.join(image_folder, img_info['file_name'])
    #  
    img = cv2.imread(img_path)
    #  numpy  
    img = np.array(img)
    
    # bounding box  
    img_annotations = [ann for ann in annotations if ann['image_id'] == img_id]
    bboxes = [ann['bbox'] for ann in img_annotations]
    category_ids = [ann['category_id'] for ann in img_annotations]
    
    #   bounding box 
    transformed = transform(image=img, bboxes=bboxes, category_id=category_ids)
    resized_img = transformed['image']
    resized_bboxes = transformed['bboxes']
    resized_category_ids = transformed['category_id']

    cv2.imwrite(img_path, resized_img)
    
    
    ############################################
    img_info['width'] = resized_img.shape[1]
    img_info['height'] = resized_img.shape[0]
    
    print(image_info['width'])

    # bounding box  
    for ann, bbox in zip(img_annotations, resized_bboxes):
        ann['bbox'] = list(bbox)

# # JSON  
```
